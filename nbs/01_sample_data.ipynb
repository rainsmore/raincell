```python
#| export
#| hide
import xarray as xr

from raincell.config import SAMPLE_DATA_DIR
```

```python
#| export
def open_cml_sample() -> xr.Dataset:
    """Open Sample Commercial Microwave Link (CML) data.""" 
    return xr.load_dataset(SAMPLE_DATA_DIR / "cml_douala.nc")
```

```python
open_cml_sample()
```

```python
#| export
def open_gauge_sample() -> xr.Dataset:
    """Open Sample rain gauge data.""" 
    return xr.load_dataset(SAMPLE_DATA_DIR / "gauges_douala.nc")
```

```python
open_gauge_sample()
```
